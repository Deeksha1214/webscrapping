```python
import requests
import pandas
from bs4 import BeautifulSoup
```

```python
url=("https://www.google.com/search?q=apple+laptop+price&rlz=1C1CHBF_enIN927IN927&sxsrf=ALeKk03BU7Yy5bzCwO0yUyxJy849SKUWPw:1614099573962&source=lnms&tbm=shop&sa=X&ved=2ahUKEwinqvO9vYDvAhXTzjgGHbQmCJoQ_AUoAXoECBQQAw&biw=811&bih=583")
```

```python
page=requests.get(url)
page.text
```

```python
soup=BeautifulSoup(page.text,'html.parser')
soup
```

```python
name=soup.find_all(class_="sh-np__product-title translate-content")
name
```

```python
lapi_name=[]
for i in name:
    lapi_name.append(i.text)
```

```python
lapi_name
```

```python
price=soup.find_all(class_="T14wmb")
price
```

```python
lapi_price=[]
for j in price:
    lapi_price.append(j.text)
```

```python
lapi_price
```

```python
website=soup.find_all(class_="E5ocAb")
```

```python
website
```

```python
lapi_website=[]
for k in website:
    lapi_website.append(k.text)
```

```python
lapi_website
```

```python
final_data=[]
for i,j,k in zip(lapi_name,lapi_price,lapi_website):
    final_data.append({'name':i,'price':j,'website':k})
```

```python
data=pandas.DataFrame(final_data)
data
```
